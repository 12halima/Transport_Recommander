```python
import os
import shutil


# ğŸ“ Dossier source contenant tous tes GTFS tÃ©lÃ©chargÃ©s
SOURCE_DIR = "/content/drive/MyDrive/Transport"

# ğŸ“ Dossier cible oÃ¹ on met seulement les GTFS valides
TARGET_DIR = "/content/drive/MyDrive/GTFS_VALIDES"

```

```python
# ğŸ“Œ Fichiers obligatoires
REQUIRED_FILES = {
    "agency.txt",
    "stops.txt",
    "routes.txt",
    "trips.txt",
    "stop_times.txt",
    "calendar.txt"
}

# ğŸ”§ CrÃ©er le dossier cible si n'existe pas
os.makedirs(TARGET_DIR, exist_ok=True)

print("ğŸ“‚ Configuration terminÃ©e !")
```

```python
def is_gtfs_valid(folder_path):
    """
    VÃ©rifie si un dossier GTFS contient tous les fichiers obligatoires.
    """
    files = set(os.listdir(folder_path))
    return REQUIRED_FILES.issubset(files)

print("ğŸ” Fonction de validation GTFS prÃªte.")
```

```python
# ğŸš€ Parcourir tous les dossiers dans Transport
gtfs_folders = sorted([
    f for f in os.listdir(SOURCE_DIR)
    if os.path.isdir(os.path.join(SOURCE_DIR, f))
])

print("ğŸ” VÃ©rification des dossiers GTFS...\n")

valid_count = 0

for folder in gtfs_folders:
    folder_path = os.path.join(SOURCE_DIR, folder)

    if is_gtfs_valid(folder_path):
        print(f"âœ… Valide : {folder}")

        # copie dans GTFS_VALIDES
        dest_path = os.path.join(TARGET_DIR, folder)
        shutil.copytree(folder_path, dest_path, dirs_exist_ok=True)

        valid_count += 1
    else:
        print(f"âŒ Incomplet : {folder}")

print("\n----------------------------------")
print(f"âœ” Total dossiers valides copiÃ©s : {valid_count}")
print(f"ğŸ“ Dossier final des GTFS valides : {TARGET_DIR}")
```

```python
# ğŸ“Œ Colonnes obligatoires pour chaque fichier

REQUIRED_COLUMNS = {
    "agency.txt": ["agency_id", "agency_name", "agency_timezone"],
    "stops.txt": ["stop_id", "stop_name", "stop_lat", "stop_lon"],
    "routes.txt": ["route_id", "route_short_name", "route_long_name", "route_type"],
    "trips.txt": ["route_id", "service_id", "trip_id"],
    "stop_times.txt": ["trip_id", "arrival_time", "departure_time", "stop_id", "stop_sequence"],
    "calendar.txt": ["service_id", "monday", "tuesday", "wednesday", "thursday",
                     "friday", "saturday", "sunday", "start_date", "end_date"]
}

# ğŸ”§ CrÃ©er le dossier cible si n'existe pas
os.makedirs(TARGET_DIR, exist_ok=True)
print("ğŸ“‚ Configuration terminÃ©e !")
```

```python
# ğŸ”¹ Fonction pour vÃ©rifier les colonnes obligatoires
def has_required_columns(folder_path):
    for file, columns in REQUIRED_COLUMNS.items():
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
        except Exception as e:
            print(f"âš ï¸ Impossible de lire {file_path}: {e}")
            return False
        missing_cols = [c for c in columns if c not in df.columns]
        if missing_cols:
            print(f"âŒ Colonnes manquantes dans {file_path}: {missing_cols}")
            return False
    return True
```

```python
import pandas as pd
# ğŸš€ Parcourir tous les dossiers GTFS
gtfs_folders = sorted([
    f for f in os.listdir(SOURCE_DIR)
    if os.path.isdir(os.path.join(SOURCE_DIR, f))
])

print("ğŸ” VÃ©rification des dossiers GTFS...\n")

valid_count = 0
removed_count = 0

for folder in gtfs_folders:
    folder_path = os.path.join(SOURCE_DIR, folder)

    if not has_required_columns(folder_path):
        print(f"âŒ Incomplet (colonnes manquantes) : {folder}")
        removed_count += 1
        continue

    # Copier dans le dossier clean
    dest_path = os.path.join(TARGET_DIR, folder)
    shutil.copytree(folder_path, dest_path, dirs_exist_ok=True)
    print(f"âœ… Valide : {folder}")
    valid_count += 1

print("\n----------------------------------")
print(f"âœ” Total dossiers valides copiÃ©s : {valid_count}")
print(f"âŒ Total dossiers supprimÃ©s pour non-conformitÃ© : {removed_count}")
print(f"ğŸ“ Dossier final des GTFS valides : {TARGET_DIR}")
```

```python
import pandas as pd

# Colonnes obligatoires
mandatory_cols = ["stop_id", "stop_name", "stop_lat", "stop_lon"]

# Fonction pour compter les "vraies" valeurs manquantes
def count_missing(series):
    # Convertir en str pour dÃ©tecter les espaces
    return series.isna().sum() + (series.astype(str).str.strip() == "").sum()

# Appliquer la fonction Ã  chaque colonne obligatoire
missing_summary = {col: count_missing(stops_df[col]) for col in mandatory_cols}

# Afficher le rÃ©sultat
print("Nombre total de valeurs manquantes, NaN ou vides par colonne obligatoire :\n")
for col, count in missing_summary.items():
    print(f"{col} : {count}")
```

```python
import pandas as pd

# Pourcentage limite pour supprimer la colonne
threshold = 0.7

# Calculer le pourcentage de valeurs manquantes (NaN ou vides) par colonne
def missing_percentage(series):
    total = len(series)
    missing = series.isna().sum() + (series.astype(str).str.strip() == "").sum()
    return missing / total

# Colonnes Ã  supprimer si plus de 70 % manquantes, sauf stop_timezone
cols_to_drop = [col for col in stops_df.columns
                if col != "stop_timezone" and missing_percentage(stops_df[col]) > threshold]

# Supprimer les colonnes
stops_df.drop(columns=cols_to_drop, inplace=True)
print(f"Colonnes supprimÃ©es (>70% manquantes) : {cols_to_drop}")

# Remplacer les NaN ou vides dans stop_timezone par 'Europe/Madrid'
stops_df["stop_timezone"] = stops_df["stop_timezone"].replace(to_replace=[None, "", "nan", "NaN"], value="Europe/Madrid")
stops_df["stop_timezone"].fillna("Europe/Madrid", inplace=True)

print("âœ… Colonnes nettoyÃ©es avec succÃ¨s")
```
