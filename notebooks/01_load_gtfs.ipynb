````python
def is_gtfs_valid(folder_path):
    """
    V√©rifie si un dossier GTFS contient tous les fichiers obligatoires.
    """
    files = set(os.listdir(folder_path))
    return REQUIRED_FILES.issubset(files)

print("üîç Fonction de validation GTFS pr√™te.")
````

````python
# üöÄ Parcourir tous les dossiers dans Transport
gtfs_folders = sorted([
    f for f in os.listdir(SOURCE_DIR)
    if os.path.isdir(os.path.join(SOURCE_DIR, f))
])

print("üîç V√©rification des dossiers GTFS...\n")

valid_count = 0

for folder in gtfs_folders:
    folder_path = os.path.join(SOURCE_DIR, folder)

    if is_gtfs_valid(folder_path):
        print(f"‚úÖ Valide : {folder}")

        # copie dans GTFS_VALIDES
        dest_path = os.path.join(TARGET_DIR, folder)
        shutil.copytree(folder_path, dest_path, dirs_exist_ok=True)

        valid_count += 1
    else:
        print(f"‚ùå Incomplet : {folder}")

print("\n----------------------------------")
print(f"‚úî Total dossiers valides copi√©s : {valid_count}")
print(f"üìÅ Dossier final des GTFS valides : {TARGET_DIR}")
````

````python
# üîπ Fonction pour v√©rifier les colonnes obligatoires
def has_required_columns(folder_path):
    for file, columns in REQUIRED_COLUMNS.items():
        file_path = os.path.join(folder_path, file)
        try:
            df = pd.read_csv(file_path, encoding='utf-8', on_bad_lines='skip')
        except Exception as e:
            print(f"‚ö†Ô∏è Impossible de lire {file_path}: {e}")
            return False
        missing_cols = [c for c in columns if c not in df.columns]
        if missing_cols:
            print(f"‚ùå Colonnes manquantes dans {file_path}: {missing_cols}")
            return False
    return True
````

````python
import os
import pandas as pd

def global_trips_stats(root_folder):
    """
    Lit tous les fichiers 'trips.txt' dans tous les sous-dossiers
    et g√©n√®re un r√©sum√© global sur les colonnes :
    - total de lignes
    - total de NaN
    - pourcentage de NaN
    """

    all_dfs = []

    # Parcours des sous-dossiers
    for dirpath, _, filenames in os.walk(root_folder):
        if "trips.txt" in filenames:
            file_path = os.path.join(dirpath, "trips.txt")

            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")
                all_dfs.append(df)
            except Exception as e:
                print(f"Erreur lecture {file_path}: {e}")

    # Aucun fichier trouv√©
    if not all_dfs:
        print("Aucun fichier trips.txt trouv√©.")
        return

    # Fusion globale
    global_df = pd.concat(all_dfs, ignore_index=True)

    print(f"\n=== TOTAL GLOBAL TRIPS ===")
    print(f"Nombre total de lignes fusionn√©es : {len(global_df):,}")

    # R√©sum√© NaN par colonne
    print("\n=== STATISTIQUES DES COLONNES ===")
    for col in global_df.columns:
        nan_count = global_df[col].isna().sum() + (global_df[col] == "").sum()
        total = len(global_df)
        percent = nan_count / total * 100

        print(f"{col:25s} -> NaN: {nan_count:,} / {total:,}  ({percent:.2f}%)")

    return global_df


# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
global_trips_stats(root_folder)
````

````python
import os
import pandas as pd


def clean_trips_df(df):
    """Nettoyage complet conforme GTFS pour un DataFrame trips."""

    # Tout en string
    df = df.astype(str).replace("Nan", pd.NA)

    # Colonnes obligatoires normalis√©es
    for col in ["route_id", "service_id", "trip_id"]:
        if col in df.columns:
            df[col] = df[col].str.strip().str.lower()

    # trip_headsign
    if "trip_headsign" in df.columns:
        df["trip_headsign"] = (
            df["trip_headsign"]
            .fillna("NaN")
            .str.strip()
            .str.lower()
        )

    # shape_id
    if "shape_id" in df.columns:
        df["shape_id"] = (
            df["shape_id"]
            .replace("", pd.NA)
            .str.strip()
            .str.lower()
        )

    # direction_id (GTFS: 0, 1)
    if "direction_id" in df.columns:
        df["direction_id"] = (
            df["direction_id"]
            .replace("", pd.NA)
            .fillna("NaN")
        )


    # wheelchair_accessible (GTFS: 0,1,2)
    if "wheelchair_accessible" in df.columns:
        df["wheelchair_accessible"] = (
            df["wheelchair_accessible"]
            .replace("", pd.NA)
        )
        df["wheelchair_accessible"] = df["wheelchair_accessible"].apply(
            lambda x: x if x in ["0", "1", "2"] else "0"
        )

    return df



def clean_all_trips(root_folder):
    """
    Parcourt tous les sous-dossiers, nettoie chaque trips_clean.txt
    puis √©crase le fichier correctement.
    """

    for dirpath, _, filenames in os.walk(root_folder):
        if "trips_clean.txt" in filenames:

            file_path = os.path.join(dirpath, "trips_clean.txt")

            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")
            except Exception as e:
                print(f"Erreur lecture {file_path}: {e}")
                continue

            # Nettoyage
            df = clean_trips_df(df)

            # Sauvegarde dans le m√™me fichier
            df.to_csv(file_path, index=False, encoding="utf-8")

            print(f"‚úî Fichier nettoy√© : {file_path}")



# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
clean_all_trips(root_folder)
````

````python
import os
import pandas as pd

def remove_logical_duplicates(root_folder):
    """
    Parcourt tous les sous-dossiers, d√©tecte et supprime les doublons logiques
    dans trips_clean.txt (ignore trip_id), puis affiche le nombre de colonnes
    et √©crase le fichier nettoy√©.
    """
    for dirpath, _, filenames in os.walk(root_folder):
        if "trips_clean.txt" in filenames:
            file_path = os.path.join(dirpath, "trips_clean.txt")
            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")
            except Exception as e:
                print(f"Erreur lecture {file_path}: {e}")
                continue

            print(f"\n=== Traitement de {file_path} ===")


            # Colonnes √† comparer pour les doublons (tout sauf trip_id)
            cols_to_check = [col for col in df.columns if col != "trip_id"]

            # D√©tecter doublons
            duplicates_mask = df.duplicated(subset=cols_to_check, keep="first")
            num_duplicates = duplicates_mask.sum()
            print(f"Nombre de doublons logiques d√©tect√©s : {num_duplicates}")

            # Supprimer doublons logiques
            df = df.drop_duplicates(subset=cols_to_check, keep="first")
            print(f"Nombre de lignes apr√®s suppression : {len(df)}")


            # Sauvegarder le fichier nettoy√©
            df.to_csv(file_path, index=False, encoding="utf-8")
            print(f"‚úî Fichier nettoy√© et doublons supprim√©s : {file_path}")

# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
remove_logical_duplicates(root_folder)
````

````python
import os
import pandas as pd
import re
import unicodedata

def remove_accents(text):
    """Supprime les accents et caract√®res sp√©ciaux d'une cha√Æne."""
    if pd.isna(text):
        return text
    text = str(text)
    text = unicodedata.normalize('NFKD', text)
    text = ''.join(c for c in text if not unicodedata.combining(c))
    return text

def clean_routes_df(df):
    """Nettoyage complet conforme GTFS pour un DataFrame routes."""

    # Tout en string
    df = df.astype(str).replace(["", "nan", "NaN", "None"], pd.NA)

    # Colonnes obligatoires normalis√©es + suppression des accents
    for col in ["route_id", "route_short_name", "route_long_name", "agency_id"]:
        if col in df.columns:
            df[col] = df[col].str.strip().str.lower().apply(remove_accents)

    # route_type : doit √™tre entre 0 et 7
    if "route_type" in df.columns:
        valid_types = {str(i) for i in range(8)}
        df["route_type"] = df["route_type"].str.strip()
        df["route_type"] = df["route_type"].apply(lambda x: x if x in valid_types else pd.NA)
        df["route_type"] = df["route_type"].astype("Int64")  # nullable int

    # Couleurs : route_color et route_text_color au format hex valide
    hex_pattern = re.compile(r"^[0-9a-f]{6}$")
    for col in ["route_color", "route_text_color"]:
        if col in df.columns:
            df[col] = df[col].str.strip().apply(lambda x: x if hex_pattern.match(str(x)) else pd.NA)

    # Supprimer les lignes o√π les colonnes obligatoires ne sont pas remplies
    df = df[
        df["route_id"].notna() &
        df["route_type"].notna() &
        (df["route_short_name"].notna() | df["route_long_name"].notna())
    ]

    return df

def clean_all_routes(root_folder):
    """
    Parcourt tous les sous-dossiers, nettoie chaque routes.txt
    et cr√©e un fichier routes_clean.txt.
    """

    for dirpath, _, filenames in os.walk(root_folder):
        if "routes_clean.txt" in filenames:

            file_path = os.path.join(dirpath, "routes_clean.txt")


            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")
            except Exception as e:
                print(f"Erreur lecture {file_path}: {e}")
                continue

            # Nettoyage
            df = clean_routes_df(df)

            # Sauvegarde dans routes_clean.txt
            df.to_csv(file_path, index=False, encoding="utf-8")

            print(f"‚úî Fichier nettoy√© cr√©√© : {file_path}")

# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
clean_all_routes(root_folder)
````

````python
import os
import pandas as pd

def global_calendar_stats(root_folder):

    all_dfs = []

    # Parcours des sous-dossiers
    for dirpath, _, filenames in os.walk(root_folder):
        if "calendar.txt" in filenames:
            file_path = os.path.join(dirpath, "calendar.txt")

            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")
                all_dfs.append(df)
            except Exception as e:
                print(f"Erreur lecture {file_path}: {e}")

    # Aucun fichier trouv√©
    if not all_dfs:
        print("Aucun fichier calendar.txt trouv√©.")
        return

    # Fusion globale
    global_df = pd.concat(all_dfs, ignore_index=True)

    print(f"\n=== TOTAL GLOBAL calendar ===")
    print(f"Nombre total de lignes fusionn√©es : {len(global_df):,}")

    print("\n=== STATISTIQUES DES COLONNES ===")
    for col in global_df.columns:
        nan_count = global_df[col].isna().sum() + (global_df[col] == "").sum()
        total = len(global_df)
        percent = nan_count / total * 100

        print(f"{col:25s} -> NaN: {nan_count:,} / {total:,}  ({percent:.2f}%)")

    return global_df


# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
global_calendar_stats(root_folder)
````

````python
import os
import pandas as pd

# Colonnes √† supprimer pour clean_calendar.txt
COLUMNS_TO_DELETE_CALENDAR = [
    "name",
    "service_name",
]

def clean_calendar(root_folder):
    total_files = 0

    for dirpath, _, filenames in os.walk(root_folder):
        if "calendar.txt" in filenames:

            file_path = os.path.join(dirpath, "calendar.txt")
            print(f"\n=== Traitement de {file_path} ===")

            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")

                # Colonnes pr√©sentes uniquement
                drop_cols = [c for c in COLUMNS_TO_DELETE_CALENDAR if c in df.columns]
                df = df.drop(columns=drop_cols, errors="ignore")

                # Nouveau fichier calendar_clean.txt
                new_path = os.path.join(dirpath, "calendar_clean.txt")
                df.to_csv(new_path, index=False, encoding="utf-8")

                print(f"Colonnes supprim√©es : {drop_cols}")
                print(f"Fichier cr√©√© : {new_path}")

                total_files += 1

            except Exception as e:
                print(f"Erreur dans {file_path} : {e}")

    print(f"\n=== R√©sum√© ===\nTotal fichiers trait√©s : {total_files}")


# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
clean_calendar(root_folder)
````

````python
import os
import pandas as pd

def clean_calendar_df(df):
    """Nettoyage complet conforme GTFS pour un DataFrame calendar_clean.txt."""

    # Colonnes obligatoires
    required_columns = [
        "service_id",
        "monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday",
        "start_date", "end_date"
    ]

    # Forcer tout en string au d√©part
    df = df.astype(str).replace(["", "nan", "NaN", "None"], pd.NA)

    # Supprimer les lignes o√π une colonne obligatoire est vide
    df = df.dropna(subset=required_columns)

    # Nettoyer et convertir les jours en int (0 ou 1)
    for day in ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]:
        df[day] = df[day].str.strip().apply(lambda x: 1 if x in ["1", "1.0"] else 0).astype("Int64")

    # Normaliser service_id
    df["service_id"] = df["service_id"].str.strip().str.lower()

    # Convertir dates en int YYYYMMDD
    for col in ["start_date", "end_date"]:
        df[col] = df[col].str.strip()
        df[col] = pd.to_numeric(df[col], errors="coerce").astype("Int64")

    # Supprimer les lignes o√π la conversion a √©chou√©
    df = df.dropna(subset=["start_date", "end_date"])

    return df

def clean_all_calendar(root_folder):
    """Parcourt tous les sous-dossiers, nettoie chaque calendar_clean.txt."""

    for dirpath, _, filenames in os.walk(root_folder):
        if "calendar_clean.txt" in filenames:
            file_path = os.path.join(dirpath, "calendar_clean.txt")


            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")
            except Exception as e:
                print(f"Erreur lecture {file_path}: {e}")
                continue

            # Nettoyage
            df = clean_calendar_df(df)

            # Sauvegarde
            df.to_csv(file_path, index=False, encoding="utf-8")
            print(f"‚úî Fichier nettoy√© cr√©√© : {file_path}")


# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
clean_all_calendar(root_folder)
````

````python
import os
import pandas as pd

def calendar_duplicates_stats(root_folder):
    """
    Parcourt tous les sous-dossiers, lit tous les calendar_clean.txt,
    et affiche le nombre total de lignes et le nombre de doublons (lignes identiques).
    """

    all_dfs = []

    # Parcours des sous-dossiers
    for dirpath, _, filenames in os.walk(root_folder):
        if "calendar_clean.txt" in filenames:
            file_path = os.path.join(dirpath, "calendar_clean.txt")
            try:
                df = pd.read_csv(file_path, dtype=str, encoding="utf-8")
                all_dfs.append(df)
            except Exception as e:
                print(f"Erreur lecture {file_path}: {e}")

    # Aucun fichier trouv√©
    if not all_dfs:
        print("Aucun fichier calendar_clean.txt trouv√©.")
        return

    # Concat√©nation de tous les fichiers
    global_df = pd.concat(all_dfs, ignore_index=True)

    # Calcul des doublons (toute la ligne identique)
    duplicated_rows = global_df.duplicated(keep=False)
    n_total = len(global_df)
    n_duplicates = duplicated_rows.sum()

    # Pr√©parer un DataFrame pour l'affichage
    stats_df = pd.DataFrame({
        "Total_lines": [n_total],
        "Duplicated_lines": [n_duplicates]
    })

    return stats_df, global_df[duplicated_rows]

# Ex√©cution
root_folder = "/content/drive/MyDrive/GTFS_CLEAN"
stats_df, duplicated_df = calendar_duplicates_stats(root_folder)

print("=== Statistiques globales des fichiers calendar_clean.txt ===")
print(stats_df)

if not duplicated_df.empty:
    print("\n=== Aper√ßu des lignes dupliqu√©es ===")
    print(duplicated_df.head(10))  # Affiche les 10 premi√®res lignes dupliqu√©es
````
